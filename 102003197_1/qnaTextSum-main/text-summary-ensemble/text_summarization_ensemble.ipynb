```
import pandas as pd
import pycaret
from pycaret.utils import version
version()
```

```
df=pd.read_csv("/content/output.csv")
df
```

```
from pycaret.classification import *

cla=setup(session_id=123,silent=True,data_split_shuffle=False,data=df,target="label",normalize=True,normalize_method="zscore",
          transformation=True,transformation_method="yeo-johnson")
compare_models(fold=12,sort='Accuracy')

```

```

lr = create_model('lr')
rf = create_model('rf')
ridge = create_model('ridge')
blender_specific = blend_models(estimator_list = [lr,rf,ridge], method = 'hard')
```

```
save_model(blender_specific, 'blender')
```
